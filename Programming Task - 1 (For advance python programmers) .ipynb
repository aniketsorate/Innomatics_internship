## Question1:

https://leetcode.com/problems/running-sum-of-1d-array/submissions/1848317595/

```python
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        running_sum = []
        sum = 0
        for i in nums:
            sum = sum + i
            running_sum.append(sum)
        return running_sum    
```

## Question2:
https://leetcode.com/problems/shuffle-the-array/submissions/1848327413/

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        shuffled_array = []
        for i in range(n):
            shuffled_array.append(nums[i])
            shuffled_array.append(nums[i+n])
        return shuffled_array
        
```

## Question3:

https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_candies = max(candies)
        result = []
        for i in candies:
            if i+extraCandies >= max_candies:
                result.append(True)
            else:
                result.append(False)
        return result
```
